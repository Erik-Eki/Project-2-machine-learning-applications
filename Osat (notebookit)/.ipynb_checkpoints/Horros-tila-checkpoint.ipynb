# Importataan kaikki tarvittava kirjastot moduulista

```python
import sys
sys.path.insert(0, './Moduulit')

from kirjastot import *
```

# Datan haku

```python
df = database_query('Select * FROM SensorData')
```

# Putsaus
- Ottaa pois kaupan aukioloaikojen ulkopuoliset ajat poijjes -> huomio analyysis

```python
df = clean_dataframe(df)
df.head()
```

# Otetaan Timestampistä difference sekuntteina

```python
df['timediff'] = df.timestamp.diff()
df.timediff = df.timediff.dt.seconds
df = df.fillna(0)
df.head()
```

# Plotataan tästä distplotti
## Nähdään että siellä on jonkun verran outliereitä. koskapa suurin osa arvoista on 0 seudulla

```python
import seaborn as sns
sns.distplot(df.timediff)
```

# Otetaan kaikkien yli 10 sekunnin differensien määrä
## määräksi saatiin about 16% koko setistä

```python
df.loc[df.timediff >= 10].shape[0] / len(df)
```

# Otetaan kaikki alle 60 sekunnin timedifferencet ja distplotataan nämä
## Nähdään että 10 sekunnin kohdalla on paljon arvoja -> mahdollinen horrostila
- Eli dataa tulee about 10 sekunnin välein


```python
df1 = df.loc[df.timediff < 60]
sns.distplot(df1.timediff)
```

# Katsotaan miltä x ja y jakauma näyttää yli ja alle 10 sekunnin arvoille
## Y:n arvot ovat piikkinä lähellä 0:llaa joten vaihtelua ei juurikaan ole -> Nodet eivät liiku

```python
sns.distplot(df1.loc[df1.timediff >= 10].x)
sns.distplot(df1.loc[df1.timediff >= 10].y).set_title('X ja Y jakauma >10 sekunnin timedifferenceille')
plt.legend(['x','y'])
plt.xlabel('Koordinaatit')

plt.xlim(-1000, 5000)

xvar1 = df1.loc[df1.timediff >= 10].x.var()
yvar1 = df1.loc[df1.timediff >= 10].y.var()

print('x-mean:',df1.loc[df1.timediff >= 10].x.mean())
print('x-var:',xvar1)
print('\ny-mean:',df1.loc[df1.timediff >= 10].y.mean())
print('y-var:',yvar1)
```

```python
sns.distplot(df1.loc[df1.timediff < 10].x)
sns.distplot(df1.loc[df1.timediff < 10].y).set_title('X ja Y jakauma <10 sekunnin timedifferenceille')
plt.ylim(0, 0.002)
plt.xlim(-1000, 5000)
plt.legend(['x','y'])
plt.xlabel('Koordinaatit')

xvar2 = df1.loc[df1.timediff < 10].x.var()
yvar2 = df1.loc[df1.timediff < 10].y.var()

print('x-mean:',df1.loc[df1.timediff < 10].x.mean())
print('x-var:',xvar2)
print('\ny-mean:',df1.loc[df1.timediff < 10].y.mean())
print('y-var:',yvar2)
```

# Nähdään että arvot ovat paljon laajemmalla alueella kuin yli 10 sekunnin timediff arvoilla
## mutta varianssia x ja y arvoilla ei kuitenkaan näytä olevan

```python
print('Alle 10 sekunnin arvot - yli 10 sekunnin arvot:')
print('Näiden x vaihtelu ero:',(xvar2/xvar1).round(1))
print('Näiden y vaihtelu ero:',(yvar2/yvar1).round(1))
```

# Horrostilaan meneminen, aijemmin nähtiin että 10 sekunnin seudulla oli timediff arvoja joten nyt selvitetään milloin horrostila saavutetaan
## Tehdään lista johon kasataan timediff summat kun timediff saavuttaa >10 sekunnin rajan

```python
sumofdiff = []
tempdiff = 0
for diff in df1.timediff:
    #print(diff)
    if diff >= 10:
        sumofdiff.append(tempdiff)
        tempdiff = 0
    else:
        tempdiff = tempdiff+diff
df_sumofdiff = pd.DataFrame()
df_sumofdiff['diff'] = sumofdiff
```

# Distplotista nähdään että arvot sijoittuvat 7 - 10 seutuun
## Suurin arvo on 10 seudulla, joka meinaa että >=10 timediff saavutetaan kun sumofdiff on 10 seudulla


```python
sns.distplot(df_sumofdiff, kde_kws={'bw': 0.1})
plt.xlabel('Sum of timediff')
plt.title('Horros')
```

# Leikataan hieman outliereitä ja sama plotti
## Konkluusio: Horrostilaan mennään kun ollaan oltu paikoillaan about 10 sekunttia.

```python
sns.distplot(df_sumofdiff.loc[df_sumofdiff['diff'] < 20], kde_kws={'bw': 0.1})
plt.xlabel('Sum of timediff')
plt.title('Horros')
```
