# Benchmark

```python
import matplotlib.pyplot as plt
```

```python
db = ['MariaDB', 'CockroachDB', 'TimescaleDB', 'InfluxDB', 'YugabyteDB']

# Arvot db labelien järjestyksessä
readtime = [130,17,35,29,12]
writetime = [80,11,50,29,70]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax1 = ax.bar(db,readtime,color='b')
ax2 = ax.bar(db, writetime, bottom=readtime, color='y')

# Tarkat sekunnit plottiin
for r1, r2 in zip(ax1, ax2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="center", color="white", fontsize=12, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="center", color="white", fontsize=12, fontweight="bold")

ax.legend((ax2, ax1), ('WRITE', 'READ'))
plt.ylabel("Seconds")
plt.title("READ/WRITE Speeds")
plt.show()
```
